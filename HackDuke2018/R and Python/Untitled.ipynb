```python
import pandas as pd
```

```python
dffill = pd.read_csv("C:\Users\ECEI\Documents\Book4.csv")
dfdata = pd.read_csv("C:\Users\ECEI\Documents\HDgrandtotal.csv")
```

```python
dfdata
```

```python
for i in range(len(dffill)):
    for j in range(len(dfdata)):
        if dffill['observation_date'][i] == dfdata['observation_date'][j]:
            for column in dffill:
                dffill.set_value(i, column, dfdata[column][j])
```

```python
dffill
```

```python
for i in range(len(dffill)-1):
    if dffill['observation_date'][i][0] == dffill['observation_date'][i+1][0]:
        for j in range(len(dffill.columns)-1):
            dffill.set_value(i+1, dffill.columns[j+1], dffill[dffill.columns[j+1]][i])
```
